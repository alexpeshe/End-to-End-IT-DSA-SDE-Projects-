```python
import pandas as pd

# Define the desired column order
desired_columns = [
    'Date', 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume', 
    'MA50', 'MA200', 'Previous day closing price', 
    'Change in price', 'Percent change in price', 
    'Previous day volume', 'Change in volue', 'Percent change in volume'
]

# Create an Excel writer object
with pd.ExcelWriter('Combined_Stocks_Ordered.xlsx') as writer:
    # Dictionary of DataFrames
    stock_dataframes = {
        'Apple': apple,
        'Amazon': amazon,
        'Google': google,
        'Meta': meta,
        'Microsoft': msoft,
        'NVIDIA': nvdia,
        'Tesla': tesla
    }

    # Write each DataFrame with preserved column order
    for name, df in stock_dataframes.items():
        # Check for any missing columns
        missing_columns = set(desired_columns) - set(df.columns)
        extra_columns = set(df.columns) - set(desired_columns)
        
        # Print warnings if there are discrepancies
        if missing_columns:
            print(f"Warning: {name} is missing columns: {missing_columns}")
        if extra_columns:
            print(f"Warning: {name} has extra columns: {extra_columns}")
        
        # Reindex to match desired columns exactly
        ordered_df = df.reindex(columns=desired_columns)
        
        # Write to Excel
        ordered_df.to_excel(writer, sheet_name=name, index=False)

print("Excel file with preserved column order created successfully!")
```
